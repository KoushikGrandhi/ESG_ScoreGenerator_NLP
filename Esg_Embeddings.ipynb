```python
esg_df = esg_df[["URL", "E", "S", "G"]]
```

```python
esg_df['URL'] = esg_df["URL"].str.replace("https://", "").replace("www.", "")
esg_df['URL'] = esg_df["URL"].str.replace("www.", "")
```

```python
esg_df['URL']
```

```python
model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens')
```

```python
esg_dataframe = esg_df[esg_df[["E", "S", "G"]].any(True)]
```

```python
len(esg_dataframe)
```

```python
non_esg_dataframe = esg_df[(esg_df["E"] == False) & (esg_df["S"] == False) & (esg_df["G"] == False)]
```

```python
len(non_esg_dataframe)
```

```python
223866 + 94465
```

```python
embedding_esgs = []
embeddings_non_esg=[]
```

```python
url_list = list(esg_dataframe["URL"])
for index in tqdm(range(len(url_list))):
    embedding_esgs.append(model.encode(url_list[index]))
```

```python
url_list = list(non_esg_dataframe["URL"])
for index in tqdm(range(len(url_list))):
    embeddings_non_esg.append(model.encode(url_list[index]))
```

```python
with open('new_esg_embeddings.npy', 'wb') as fp:
    np.save(fp, embedding_esgs)
```

```python
with open("not_esg_embeddings.npy",'wb') as fp:
  np.save(fp,embeddings_non_esg)
```
